# Reflection coefficients for TE and TM modes
- **Goal**: to determine the reflectance for a multilayered structure consisting of different dielectric materials of various thicknesses
    - To analyse the special case of d = lambda / 2

```python
import numpy as np
from numpy import cos, sqrt, sin, pi
import matplotlib.pyplot as py
```

```python
# initial variables

n = 1.5                          # refractive index
N_max = 100                      # num of points for plot
t = np.linspace(0, 1, N_max)     # theta arguments
theta = (np.pi / 2) * t          # angles in radians
angle_degrees = (theta/pi) * 180 # angles in degrees
```

```python
# get equations

num_TE = cos(theta) - sqrt(n**2 - sin(theta) ** 2)
den_TE = cos(theta) + sqrt(n**2 - sin(theta) ** 2)
r_TE = num_TE / den_TE

num_TM = n ** 2 * cos(theta) - sqrt(n**2 - sin(theta)**2)
den_TM = n ** 2 * cos(theta) + sqrt(n**2 - sin(theta)**2)
r_TM = num_TM / den_TM
```

```python
# plot

x_line = [0, max(angle_degrees)]
y_line = [0, 0]

py.plot(angle_degrees, r_TE);
py.plot(angle_degrees, r_TM);
py.plot(x_line, y_line, linestyle='dashed', color='black', alpha=0.3);

py.xlabel('Angle of incidence');
py.ylabel('Reflection coefficient');
py.legend(['r_TE', 'r_TM']);
```
